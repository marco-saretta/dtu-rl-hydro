# Week 0 - BESS RL Exercises

## Exercise 1 - Bellman’s Equation

**Question:**  
Write Bellman’s equation for the given BESS system and identify:
- State  
- Action  
- Reward  
- Discount  
- Transition probability  

**Assumptions/Data:**  
- BESS capacity: 1 MWh  
- Price taker  
- Price time series known  
- Discrete SOC states: 0-1 MWh in 0.1 steps (10 states total)

## Exercise 2 - Training a Policy

**Question:**  
Train a policy for the BESS using historical data and explain how you would:
- Represent the policy  
- Choose between offline and online approaches  
- Handle exploration vs. exploitation  
- Evaluate policy performance  

**Assumptions/Data:**  
- Same 1 MWh BESS setup as above  
- Access to historical electricity prices  
- Discrete actions: charge, discharge, idle
